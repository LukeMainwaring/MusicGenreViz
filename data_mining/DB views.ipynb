```python
query = '''
CREATE VIEW genre_by_year AS (
    SELECT
        ar.year AS year,
        COUNT(*) AS num_genres
    FROM artist_rank ar
    NATURAL JOIN artist_genre ag
    GROUP BY
        year
    ORDER BY 
        year
)
;
'''

with connection.cursor() as cursor:
    cursor.execute(query)
    result = cursor.fetchall()
    for x in result:
        print(x)
```

```python
query = '''
CREATE VIEW genre_by_city AS (
    SELECT
        c.city_name AS city_name,
        COUNT(*) AS num_genres
    FROM artist_city ac
    NATURAL JOIN artist_genre ag
    NATURAL JOIN city c
    GROUP BY
        city_id
    ORDER BY 
        city_id
)
;
'''

with connection.cursor() as cursor:
    cursor.execute(query)
    result = cursor.fetchall()
    for x in result:
        print(x)
```

```python
query = '''
SELECT 
    ag.genre, 
    COUNT(*)/(SELECT num_genres FROM genre_by_city WHERE city_id = {city_id}) AS percent
FROM artist_city ac
NATURAL JOIN artist_genre ag
NATURAL JOIN city c
WHERE city_id = {city_id}
GROUP BY
    genre
ORDER BY 
    percent DESC
LIMIT 10
;
'''.format(city_id='0')


with connection.cursor() as cursor:
    cursor.execute(query)
    result = cursor.fetchall()
    for x in result:
        print(x)
```
